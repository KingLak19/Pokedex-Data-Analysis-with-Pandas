```python
import pandas as pd
```

### WAYS TO READ FROM 3 DIFFERENT FILES

```python
# df = dataframe (Same as tables)

# pd.read_excel to read from MS Excel files (file extension = .xlsx)
df_excel = pd.read_excel('pokemon_data.xlsx')

# pd.read_csv(<filename>.csv, delimeter = '<separator>') to read from .txt files (tab separated value files)
df_txt = pd.read_csv('pokemon_data.txt',delimiter='\t') # \t = tab (4 space)

# pd.read_csv  to from from csv (comma-separated value) files
df = pd.read_csv('pokemon_data.csv')
```

```python
# number of rows to be passed as an argument to pd.head function [Default = 5]
df.head(7)
```

```python
# To print only the Column headings
# Each column in a dataframe is a 'series'

df.columns
```

### To print particular columns

## Selection of elements :

```python
# iloc = index-based selection (row,column)
df.iloc[20:31,1]      #19th-50th row and Columns with Index 1
```

```python
# loc =  label-based selection (row,column
df.loc[69:,['Type 1','Type 2']]
```

```python
# print(df.iloc[:10,1])
df.loc[df.Name == 'CharizardMega Charizard X']
```

-------------------------------------------------------------------------------------------------------------------------------

## Operators :

```python
# & = and
# | = or
df.loc[(df['Legendary'] == True) & (df['Generation'] == 1)]
```

```python
# printing all non-null values
df.loc[df['Type 2'].notnull()]
```

-------------------------------------------------------------------------------------------------------------------------------

 ## Making new columns :

```python
# Making changes to the Data
# Adding a new column
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']

df
```

## Renaming Columns

```python
df.columns

df.columns = ['DexNo','Name','Type1', 'Type2', 'HP', 'Attack', 'Defense', 'Sp_Atk',
       'Sp_Def', 'Speed', 'Generation', 'Legendary', 'Total']

df
```

-------------------------------------------------------------------------------------------------------------------------------

## Dropping a column :

```python
# To drop a column
df = df.drop(columns = ['Total'])

df
```

--------------------------------------------------------------------------------------------------------------------------------

## Making columns using Indexes :

### It is always better to use names for safety

```python
# Axis=1 represents horizontal Axis

df['Total'] = df.iloc[:,4:10].sum(axis = 1)

df
```

--------------------------------------------------------------------------------------------------------------------------------

## Re-arranging columns :

```python
cols = list(df.columns)

df = df[cols[0:2]+cols[4:10]+[cols[-1]]+cols[2:4]+cols[10:12]]

df
```

-----------------------------------------------------------------------------------------------------------------------------------------

## Strongest pokemon according to stats :

```python
df.loc[df.Total == df.Total.max()]
```

--------------------------------------------------------------------------------------------------------------------------------

## Weakest pokemon according to stats

```python
df.loc[df.Total == df.Total.min()]
```
